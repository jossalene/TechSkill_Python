```python
# because the data has the column containing date, we MUST tell pandas about the dates
# parse_dates tells pandas that we have a date object so it will treat it like a date
# and not like a string
shelter = pd.read_csv("DHS_Daily_Report.csv", parse_dates = ["Date of Census"])

#check the dataframe created correctly by displaying it on the screen
# .head() to show only the first 5 lines of the dataframe
shelter.head()
```

```python
# data types of each variable
shelter.dtypes
```

```python
shelter.plot(x = "Date of Census", y = "Total Individuals in Shelter")
```

```python
shelter2019 = pd.read_csv("DHS_Daily_Report2019.csv", parse_dates = ["Date of Census"])
shelter2019.plot(x = "Date of Census", y = "Total Individuals in Shelter")
```

```python
# must have y
shelter2019.plot(x = "Date of Census")
```

```python
# lift the legend so we can see x
shelter2019.plot(x = "Date of Census").legend(bbox_to_anchor = (1,1))
```

```python
# proportion of individuals who are adults in the shelter
# number of individual who are adult / total number of individuals

# shelter["Proportion of adult in the shelter"] = shelter["Total Adults in Shelter"] / shelter["Total Individuals in Shelter"] 

# OR

a = shelter["Total Adults in Shelter"]
b = shelter["Total Individuals in Shelter"]
shelter["Proportion Adults"] =  a / b 

shelter["Proportion Adults"]
```

```python
shelter.head()
```

```python
shelter.plot(x = "Date of Census", y = "Proportion Adults")
```

```python
shelter2019["Proportion Adults"] = shelter2019["Total Adults in Shelter"] / shelter2019["Total Individuals in Shelter"] 
shelter2019["Proportion Adults"]
```

```python
shelter2019.plot(x = "Date of Census", y = "Proportion Adults")
```

```python
# year : 2020
shelter["Proportion Children"] = shelter["Total Children in Shelter"] / shelter["Total Individuals in Shelter"] 

# year : 2019
shelter2019["Proportion Children"] = shelter2019["Total Children in Shelter"] / shelter2019["Total Individuals in Shelter"] 

shelter.plot(x = "Date of Census", y = "Proportion Children", title = "Proportion Children in 2020")
shelter2019.plot(x = "Date of Census", y = "Proportion Children", title = "Proportion Children in 2019")
```

```python
# year : 2020
shelter["Proportion Adults "] = shelter["Total Adults in Shelter"] / shelter["Total Individuals in Shelter"] 

# year : 2019
shelter2019["Proportion Adults"] = shelter2019["Total Adults in Shelter"] / shelter2019["Total Individuals in Shelter"] 

shelter.plot(x = "Date of Census", y = "Proportion Adults", title = "Proportion Adults in 2020")
shelter2019.plot(x = "Date of Census", y = "Proportion Adults", title = "Proportion Adults in 2019")
```
