`.loc` will be used to select rows/columns in a dataset  
`.loc` requires two parameters -- (1) index of the row(s) to be selected, (2) the column(s) to be selected.

```python
zillow.head(n=12)
```

```python
zillow.loc[10, "RegionName"]
```

```python
zillow.loc[10, ['RegionName', 'State', 'Metro', 'County']]
```

`.iloc` -- uses the row index or the column index

```python
zillow.iloc[10, 2]
```

```python
# this will return ERROR because iloc take INDEX, NOT NAME!!
zillow.iloc[10, ['RegionName', 'State', 'Metro', 'County']]
```

```python
# select a single row with values from all columns
zillow.loc[15,]
```

```python
# OR
zillow.loc[15, :]
```

```python
# select multiple rows but 1 column
zillow.loc[1000:1010,"RegionName"]
```

```python
zillow.loc[ :,"RegionName"]
```

```python
zillow.loc[:,"RegionName" : "County"]
```

```python
# get all row for column 2,3,4,5
zillow.iloc[:,2:6]  # up to 6 but not 6
```

```python
# get multiple rows but with NON-CONTIGUOUS columns
zillow.loc[5432:5450, ["RegionName", "Metro", "State"]]
```

## Filtering

`filtering` -- a way to select rows/columns based on a certain value or condition

```python
zillow["RegionName"].value_counts()
```

```python
#filtering return boolean values
#basically, it will return False if the condition is not met; True if the condition is met
#we are essentially getting back a Boolean series

Dallas_filter = zillow["RegionName"] == "Dallas"
Dallas_filter 
```

```python
# apply the filter using square brackets BUT NO QUOTATION MARKS
# it will return when filter is True

zillow[Dallas_filter]
```

```python
# in one line

zillow[zillow['RegionName'] == 'Dallas']
```

```python
# sort values
#sorting occurs in ASCENDING order by default

zillow.sort_values("Metro")
```

```python
zillow.sort_values("Metro", ascending = False)
```

```python
zillow.sort_values(by = ["State","County"])
```

```python
sorted = zillow.sort_values(by =['Metro','County','Zhvi'], \
                             ascending = [True, False, False])
sorted.head()
```

```python
price_filter = zillow["Zhvi"] >= 500000
price_filter.head()
```

```python
zillow[price_filter]
```

```python
new_price_filter = zillow["Zhvi"] > 2500000
state_filter = zillow["State"] == "NY"
```

```python
zillow[new_price_filter & state_filter]
```

```python
zillow[new_price_filter |  (zillow["State"] == "CA")]
```

and == &  
or == |

```python
zillow[new_price_filter | (zillow["State"] == "NY")].head()
```
